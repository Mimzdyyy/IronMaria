```python
response_claim = dataset.groupby(['Response','Total Claim Amount']).size().reset_index(name = 'Counts')
```

```python
f1, ax1 = plt.subplots(figsize=(8, 6))
fig1 = sns.barplot(x=response_claim["Total Claim Amount"], y = response_claim["Counts"],hue=response_claim['Response'])
ax1.set(ylim=(0,50),xticks=[10000])

ax1.set_title('The response rate by the total claim amount')
```

```python
f1, ax1 = plt.subplots(figsize=(8, 6))
fig1 = sns.scatterplot(x=response_claim["Total Claim Amount"], y = response_claim["Counts"])
ax1.set(ylim=(0,150),xticks=[100000])

ax1.set_title('The response rate by the total claim amount')
```
